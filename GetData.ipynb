```python
import requests, time, tqdm, pandas, pickle
```

```python
api = 'https://api.vk.com/method/'
key = open("text.txt")
token = key.readline()[0:-1]
```

```python
sad = requests.get(api + 'groups.search?q=грустные&count=20&v=5.85' + token).json()['response']['items']
funny = requests.get(api + 'groups.search?q=веселые&count=20&v=5.85' + token).json()['response']['items']
```

```python
data1={}
data0={}
```

```python
for i in tqdm.tqdm_notebook(sad):
    time.sleep(0.5)
    sad_gays=requests.get(api + 'users.search?count=5&group_id=' + str(i['id'])+'&v=5.85' + token).json()['response']['items']
    data1[i['id']]= sad_gays
```

```python
for i in tqdm.tqdm_notebook(funny):
    time.sleep(0.5)
    funny_gays=requests.get(api + 'users.search?count=5&group_id=' + str(i['id'])+'&v=5.85' + token).json()['response']['items']
    data0[i['id']]= funny_gays
```

```python
data11 = {'first_name': [], 'last_name': [], 'id': [], 'suicide': [], 'group': []}
```

```python
for key, val in data1.items():
    for info in val:
        data11['first_name'].append(info['first_name'])
        data11['last_name'].append(info['last_name'])
        data11['id'].append(info['id'])
        data11['suicide'].append(1)
        data11['group'].append(key)
```

```python
for key, val in data0.items():
    for info in val:
        data11['first_name'].append(info['first_name'])
        data11['last_name'].append(info['last_name'])
        data11['id'].append(info['id'])
        data11['suicide'].append(0)
        data11['group'].append(key)
```

```python
frame = pandas.DataFrame(data11)
frame
```

```python
with open('gays1.pickle', 'wb') as f:
    pickle.dump(frame, f)
```
