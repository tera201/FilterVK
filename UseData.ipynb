```python
import pickle, pandas, requests, tqdm, time
```

```python
with open('gays.pickle', 'rb') as f:
    frame = pickle.load(f)
```

```python
frame
```

```python
ids = list(frame['id'])

def chunks(lst, chunk_size):
    return [lst[i:i + chunk_size] for i in range(0, len(lst), chunk_size)]

ids = chunks(ids, 20)
```

```python
api = 'https://api.vk.com/method/'
key = open("text.txt")
token = key.readline()[0:-1]
```

```python
data = {'id': [],
    'sex': [], 'relation': [],
    'political': [], 'religion': [], 'people_main': [], 'life_main': [],
    'smoking': [], 'alchol': [], 'groups': []
}

```

```python
for uid in tqdm.tqdm_notebook(ids):
    ID=','.join(list(map(str, uid)))
    IDg=[]
    time.sleep(0.1)
    gays = requests.get(api + 'execute.infoUser' + '?id=' + ID + '&v=5.85' + token).json()
    get = gays['response'][0]
    
    for x in get:
        data['id'].append(x['id'])
        data['sex'].append(x['sex'])
        data['religion'].append(x.setdefault('personal', {}).setdefault('religion', None))
        data['relation'].append(x.setdefault('relation', None))
        data['political'].append(x.setdefault('personal', {}).setdefault('political', None))
        data['people_main'].append(x.setdefault('personal', {}).setdefault('people_main', None))
        data['life_main'].append(x.setdefault('personal', {}).setdefault('life_main', None))
        data['smoking'].append(x.setdefault('personal', {}).setdefault('smoking', None))
        data['alchol'].append(x.setdefault('personal', {}).setdefault('alcohol', None))
        IDg.append(x['id'])
    
    time.sleep(0.3)
        
    ID=','.join(list(map(str, IDg)))
    groups =requests.get(api + 'execute.infoUsers' + '?id=' + ID + '&v=5.85' + token).json()['response']
    
    #print(len(groups),len(get))
    for user in groups:
        tmp = []
        if type(user) == type(False):
            data['groups'].append([])
            continue
        for group in user['items']:
            tmp.append(group.setdefault('id', None))
        data['groups'].append(tmp)
        
    
```

```python
import pickle
with open('FullFrame.pickle', 'wb') as f:
    pickle.dump(data, f)
```
